## 1. Imports

```python
import os
import math
import random
from datetime import datetime

print('Libraries imported successfully!')
```

## 2. Variables & Basic Operations

```python
# Basic variables
name = 'Alice'
age = 30
pi = math.pi

print(f'Name: {name}')
print(f'Age: {age}')
print(f'Pi: {pi:.4f}')
```

## 3. Lists & Loops

```python
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']

print('Fruits list:')
for i, fruit in enumerate(fruits, 1):
    print(f'  {i}. {fruit.capitalize()}')
```

## 4. Functions

```python
def greet(name: str, greeting: str = 'Hello') -> str:
    """Return a greeting message."""
    return f'{greeting}, {name}! Today is {datetime.now().strftime("%A, %B %d %Y")}.'

print(greet('Alice'))
print(greet('Bob', greeting='Hi'))
```

## 5. Dictionary & Comprehensions

```python
# Dictionary
person = {
    'name': 'Alice',
    'age': 30,
    'skills': ['Python', 'Data Analysis', 'ML']
}

# List comprehension — squares of 1–10
squares = {x: x**2 for x in range(1, 11)}

print('Person:', person)
print('Squares:', squares)
```

## 6. Simple Stats (no extra libraries needed)

```python
data = [random.randint(1, 100) for _ in range(20)]

mean = sum(data) / len(data)
minimum = min(data)
maximum = max(data)

print(f'Data  : {data}')
print(f'Mean  : {mean:.2f}')
print(f'Min   : {minimum}')
print(f'Max   : {maximum}')
```
